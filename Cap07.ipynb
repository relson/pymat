## Exemplo 7.11 - Formatação dos eixos sem contorno

```python
from pylab import subplots
```

```python
x = arange(-2, 4, 0.5)
y = 3 * x
```

```python
fig, ax = subplots()
ax.plot(x, y)

ax.grid(True, which='both')

ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

ax.yaxis.tick_left()
ax.xaxis.tick_bottom()

xlabel('Variavel Independente x', horizontalalignment='right', position=(1, 10))

ylabel('Funcao y = 3x', verticalalignment='center', position=(1,0.7))
title('Nome do Grafico')
```

## Exemplo 7.16 - Formatação

```python
from pylab import subplot
```

```python
x=arange(0,4,0.5)
y=2*x
y1=3*y
y2=y**2
```

```python
subplot(121)
plot(x,y,'g^', x, y1, 'rd', x, y2, 'b*')
title('Grafico 1')

subplot(122)
yscale('log')
plot(x, y,'mv',x,y1,'wo',x,y2,'k8')
title('Grafico 2 - Eixo y Esc Log')
```

## Exemplo 7.17 - Formatação

```python
from pylab import subplots_adjust
```

```python
subplots_adjust(left=0.125,bottom=0.1,right=0.9, top=1.1,wspace=0.2,hspace=0.4)

subplot(221)
plot(x,y,'g^',x,y1,'ro',x,y2,'bs')
title('Grafico 1')

subplot(222)
yscale('log')
plot(x, y, 'k*', x, y1,'cd',x,y2,'mp')
title('Grafico 2 - Eixo y Escala Log')

subplot(223)
plot(x, y / 2, 'yh', x, 2 * y1, 'wo', x, y2 / 3, 'gx')
title('Grafico 3')

subplot(224)
xscale('log')
plot(x**2,y,'kv',x**3, y1, 'c8',x**2.5, y2/2,'b.')
title('Grafico 4 - Eixo x Escala Log')
```

## Exemplo 7.18 - Função Polinomial

```python
def f(x):
    return x**2-2*x-3
```

```python
x=arange(-5.0,7.0,0.01)
y=f(x)
```

```python
plot(x, y)

xlabel('Variavel Independente x')
ylabel('Funcao y = f(x)')
title('Funcao Quadratica y = x**2 - 2 * x - 3')
grid(True)
```

## Exemplo 7.19 - Função Trigonométrica

```python
from math import sin,pi
from pylab import *
```

```python
def f2(x):
    return 2 * sin(x/2 + pi/3)
```

```python
x = arange(0.0,6*pi+pi/3,0.01)
y = f2(x)
```

```python
plot(x, y)

xlabel('Variavel Independente x')
ylabel('Funcao y = f(x)')
title('Funcao Trigonometrica y = 2sin(x/2 + p1/3)')
grid(True)

axis([0,20,-2.5,2.5])
```

## Exemplo 7.20 - Função Exponencial

```python
from math import exp
```

```python
def exp_f(x):
    return 2 * exp(-x / 4)
```

```python
x = arange(-1,9,0.01)
y = exp_f(x)
```

```python
plot(x, y)

xlabel('Variavel Independente x')
ylabel('Funcao y = f(x)')
title('Funcao Exponencial y = 2*exp(-x/4)')
grid(True)
```

## Exemplo 7.21 - Função Logarítmica

```python
from numpy import log10,arange
```

```python
def log_f(x):
    return 3 * log10(x ** 2)
```

```python
x = arange(1.0,9.0,0.01)
y = log_f(x)
```

```python
plot(x,y)

xlabel('Variavel Independente x')
ylabel('Funcao y = f(x)')
title('Funcao Logaritmica y = 3*log10(x**2)')
grid(True)
```

## Exemplo 7.22 - Histograma

```python
from numpy import random
from pylab import hist
```

```python
random.seed(200)
media=10
desvio_padrao=3

valores=media+desvio_padrao+random.randn(3000)
```

```python
xlabel('Valores')
ylabel('Probabilidade')

title('Histograma')
axis([0,20,0,0.2])
text(1,0.17,'Media = 10, Desvio Padrao = 3')

hist(valores, 30, density=True,facecolor='orange')
```

## Exemplo 7.23 - Barras

```python
from pylab import bar
```

```python
valores=[5,16,10,22]
x=[1,2,3,4]

bar(x,valores,width=0.5,align='center')

xlabel('Valores')
ylabel('Unidade')

title('Grafico de Barras')
axis([0,5,0,30])
```

## Exemplo 7.24 - Barras - Várias Cores

```python
rotulo_barras=['Prod 1','Prod 2','Prod 3','Prod 4']
```

## Exemplo 7.27 - Número Complexos - Polar

```python
from pylab import polar,angle
```

```python
z1 = 2 - 5j
z2 = -3 + 2j
```

```python
polar([0, angle(z1)], [0, abs(z1)], marker='o')
polar([0, angle(z2)], [0, abs(z2)], marker='*')
```

## Exemplo 7.28 - Número Complexo - Retangular

```python
from pylab import axhline,axvline
```

```python
plot([0, z1.real],[0,z1.imag], 'ro-')
text(z1.real + 0.2, z1.imag + 0.2, 'z1')

plot([0,z2.real],[0,z2.imag], 'b*--')
text(z2.real + 0.2, z2.imag + 0.2, 'z2')

axis([-4,3,-7,7])

axhline(y=0,color='k')
axvline(x=0,color='k')

grid(True)
```
