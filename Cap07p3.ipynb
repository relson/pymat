```python
from math import sin,pi
from pylab import arange,plot,xlabel,ylabel,title,grid,axis
```

```python
def f2(x):
    return 2 * sin(x/2 + pi/3)
```

```python
x = arange(-5.0,7.0,0.01)
y = f2(x)
```

```python
plot(x,y)

xlabel('Variavel Independente x')
ylabel('Funcao y = f(x)')
title('Funcao Trigonometrica y = 2sin(x/2 + p1/3)')
grid(True)

axis[0,20,-2.5,2.5]
```
